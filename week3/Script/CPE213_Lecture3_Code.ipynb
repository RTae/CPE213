## Convert character to factor

```
mpg1 <- mpg %>%
  mutate_if(is.character, as.factor)
```

## Summary

```
summary(mpg)
```

```
summary(mpg1)
```

## ggplot2: Example

### Scater plot

```
p1 <- ggplot(mpg1) + 
  geom_point(aes(cty, hwy, color=class), size=2)
p2 <- p1 +
  geom_smooth(aes(cty, hwy), method='lm') + 
  scale_color_brewer(type='qual')
p2 + facet_grid(.~cyl)
```

```
ggplot(mpg1) + 
  geom_point(aes(x = hwy, cty)) 

## หาความสัมพันธ์เชิงเส้น
```

#### geom_jiter

```
ggplot(mpg1) +
  geom_jitter(aes(hwy, cty))
```

### Bar plot

#### geom_bar

```
ggplot(mpg1) + 
  geom_bar(aes(class))

# จากกราฟนี้มันจะอธิบายว่า Prob ที่จะเจอรถรุ่นไหนในท้องตลาดมากที่สุด
```

#### geom_col

```
mpg1 %>%
  group_by(class) %>%
  summarise(avg_hwy = mean(hwy)) %>%
  ggplot() +
  geom_col(aes(class, avg_hwy, fill=class))

# จากกราฟนี้มันจะอธิบายว่า รถประเภทไหน ประหยัดพลังงานมากว่ากัน (เน้นการเปรียบเทียบ)
```

#### geom_histogram

```
ggplot(mpg1) +
  geom_histogram(aes(hwy, fill=class), dotsize=0.5)
```

#### geom_density

```
ggplot(mpg1) +
  geom_density(aes(hwy))
```

#### geom_area

```
ggplot(mpg1) +
  geom_area(aes(hwy), stat='bin')
```

#### geom_dotplot

```
ggplot(mpg1) +
  geom_dotplot(aes(hwy, fill = class), dotsize=0.7)
```

### Box plot 



#### geom_boxplot

```
ggplot(mpg1) + 
  geom_boxplot(aes(class,hwy,fill=class))
```

#### geom_violin

```
ggplot(mpg1) +
  geom_violin(aes(x = class, y = hwy, fill = class))
```

### 2D Distribution

#### geom_bin2d

```
ggplot(mpg1) + 
  geom_bin2d(aes(x=displ, y=cty), bins=10) +
  scale_fill_gradient(low="#000000", high="#FF0000")
```

#### geom_density2d

```
ggplot(mpg1) + 
  geom_density2d(aes(x=displ, y=cty))
```

## Combine

```
mpg1 %>%
  group_by(class) %>%
  summarise(cty = mean(cty), hwy=mean(hwy)) %>%
  ggplot(aes(hwy, cty)) + geom_point() +
  geom_text(aes(label = class), nudge_y=-0.3) + 
  xlim(15,30)
```

## Line graph

### geon_line

```
ggplot(economics, aes(date, unemploy)) + 
  geom_line()
```

### geom_ribbon

```
ggplot(economics, aes(date, unemploy)) + 
  geom_ribbon(aes(ymin=unemploy - 900, ymax=unemploy+900), alpha=0.3) + 
  geom_line()
```

### geom_step

```
ggplot(economics, aes(date, unemploy)) + 
  geom_step(direction = "hv") + 
  xlim(as.Date(c('1/1/1970', '1/1/1980'), format="%d/%m/%Y"))
```
