# Import library

```
library("tidyverse")
```

# Load flights.csv and airline.csv to dataframe

```
df_flights <- read.csv("flights.csv")
df_airlines <- read.csv("airlines.csv")
```

```
df_flights
```

```
df_airlines
```

# Select two columns (carrier, dep_delay)

```
str(df_flights)
```

```
df_flights_select <- df_flights %>%
                        select(carrier, dep_delay)
```

```
df_flights_select
```

# Filter NA out of dep_delay

```
df_flights_select_out_NA <- df_flights_select %>%
                              drop_na(dep_delay)
```

```
df_flights_select_out_NA
```

# Use carrier as a group and calculate mean of dep_dealy

```
df_mean_delay_of_each_carrier <- df_flights_select_out_NA %>%
                                    group_by(carrier) %>%
                                    summarise(mean_delay = mean(dep_delay))
```

```
df_mean_delay_of_each_carrier
```

# Sort the wort carrier by departure delay (dep_deplay)

```
df_mean_delay_of_each_carrier_sort <- df_mean_delay_of_each_carrier %>%
                                        arrange(desc(mean_delay))
```

```
df_mean_delay_of_each_carrier_sort
```

# Join airlines data to flight data using "carrier" as a key

```
df_mean_delay_of_each_carrier_sort %>%
    left_join(df_airlines, by="carrier") %>%
    select(carrier, mean_delay, name)
```
